# regression part

```python
# so we can take 1,2,3 as our independent variables and left column as our dependent variable.
X=hr[['satisfaction_level','promotion_last_5years','salary']]
X
```

```python
Y=hr[['left']]
Y
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.3,random_state=3)
```

```python
reg=LogisticRegression()
reg
```

```python
reg.fit(X_train,Y_train)
```

```python
reg.predict([[0.7,0,1]])
```

```python
reg.score(X_test,Y_test)
```
