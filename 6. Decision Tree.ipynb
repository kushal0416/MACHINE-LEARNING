# Decision Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix


```

1. In decison tree we will deal with gini impurity and entropy
2. Any ml first we need convert them into numerical data - we can use dummy variables or label encoder

```python
# reading the salry dataset
sal=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv')
sal
```

# Finding X and Y

```python
X= sal.drop(['salary_more_then_100k'],axis=1)
X
```

```python
# converting the x columns to numerical before that we will print unique elements in columns
sal.company.unique()
```

```python
sal.job.unique()
```

```python
sal.degree.unique()
```

```python
#creating dummy variables
A=pd.get_dummies(sal[['company','job','degree']])
A
```

```python
Y= sal.salary_more_then_100k
Y
```

# performing decision tree

```python
reg= tree.DecisionTreeClassifier()
reg
```

```python
A_train,A_test,Y_train,Y_test=train_test_split(A,Y,test_size=0.2)
```

```python
A_test
```

```python
Y_test
```

```python
reg.fit(A_train,Y_train)
```

```python
Y_predict=reg.predict([[0,0,1,0,0,1,0,1]])
Y_predict
```

```python
reg.score(A_test,Y_test)
```
