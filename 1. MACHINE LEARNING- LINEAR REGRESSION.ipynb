# house prices

```python
ml=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/homeprices.csv')
ml
#there is one null element that we should pre process 
```

```python
ml.isnull().sum()
# there is one null in bedrooms column
# we need to work on that
```

```python
#ml.fillna({'bedrooms':3}) # we can manually guess a number and insert them by using dictionary
```

```python
# but if we have 1000 null values we cannot guess, so we use math functions like mean, median to replace null values
median_bedrooms=ml.bedrooms.median() # ml[bedrooms].median()
median_bedrooms
```

```python
# if the median is 3.5 flow the below code.
""" 
import math
median_bedrooms=math.floor(ml.bedrroms.median())
median_bedrooms
ml.bedrooms.fillna(median_bedrooms)
"""
```

```python
ml.fillna({'bedrooms':median_bedrooms},inplace=True)
ml
# now that we have our dataset we can perform multi linear regression.
```

```python
reg=LinearRegression()
reg.fit(ml[['area','bedrooms','age']],ml.price)
# syntax - reg.fit(independent variable,target variable)
```

```python
reg.predict([[3500,5,3]])
```

# hiring data set- salary prediction

```python
sp=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/Exercise/hiring.csv')
sp
```

```python
sp.isnull().sum()
# there are 2 nulls in experience and 1 in test_score.
```

```python
# data preprocessing -1
sp.experience.fillna(method='bfill',inplace=True)
sp# in this step we have filled NaN of experience using backfill.
```

```python
#data preprocessing -2 : test_score median
import math
test_score_median=sp['test_score(out of 10)'].mean()
test_score_median
# now the median is in float type we will convert to integer
```

```python
a=round(test_score_median)
a
```

```python
sp['test_score(out of 10)']=sp['test_score(out of 10)'].fillna(a)# remember the syntax it is useful
#when the attribute is multi character.
# syntax - df['attribute']=df['attribute'].function()
sp
```

```python
# the final step of preprocessing is to convert experience data into numerical data
#CONVERTING STRING DATA TO NUMBERS

from word2number import w2n

sp.experience=sp.experience.apply(w2n.word_to_num)
sp
```

```python
reg=LinearRegression()
reg.fit(sp[['experience','test_score(out of 10)','interview_score(out of 10)']],sp['salary($)'])
```

```python
reg.predict([[2,9,6]])
```
